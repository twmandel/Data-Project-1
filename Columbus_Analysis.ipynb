```python
# Create DataFrame with Reviews greater than 35

air_bnb_columbus_reviews = air_bnb_columbus.loc[air_bnb_columbus["number_of_reviews"] > 35]
air_bnb_columbus_reviews
```

```python
# Group New DataFrame by Bedrooms/Room Type and Average of Each of the 7 Scores

types_bedrooms_accuracy = air_bnb_columbus_reviews.groupby(['bedrooms','room_type'])['review_scores_accuracy'].mean()
types_bedrooms_cleanliness = air_bnb_columbus_reviews.groupby(['bedrooms','room_type'])['review_scores_cleanliness'].mean()
types_bedrooms_value = air_bnb_columbus_reviews.groupby(['bedrooms','room_type'])['review_scores_value'].mean()
types_bedrooms_communication = air_bnb_columbus_reviews.groupby(['bedrooms','room_type'])['review_scores_communication'].mean()
types_bedrooms_location = air_bnb_columbus_reviews.groupby(['bedrooms','room_type'])['review_scores_location'].mean()
types_bedrooms_check_in = air_bnb_columbus_reviews.groupby(['bedrooms','room_type'])['review_scores_checkin'].mean()
types_bedrooms_ratings = air_bnb_columbus_reviews.groupby(['bedrooms','room_type'])['review_scores_rating'].mean()

# Convert grouped data into Dataframes

types_bedrooms_accuracy_df = pd.DataFrame(data=types_bedrooms_accuracy)
types_bedrooms_cleanliness_df = pd.DataFrame(data=types_bedrooms_cleanliness)
types_bedrooms_value_df = pd.DataFrame(data=types_bedrooms_value)
types_bedrooms_communication_df = pd.DataFrame(data=types_bedrooms_communication)
types_bedrooms_check_in_df = pd.DataFrame(data=types_bedrooms_check_in)
types_bedrooms_location_df = pd.DataFrame(data=types_bedrooms_location)
types_bedrooms_ratings_df = pd.DataFrame(data=types_bedrooms_ratings)

# Merge Dataframes

first_merge = pd.merge(types_bedrooms_accuracy_df,types_bedrooms_cleanliness_df, on=['bedrooms','room_type']).merge(types_bedrooms_value_df,on=['bedrooms','room_type'])
second_merge = pd.merge(types_bedrooms_communication_df,types_bedrooms_check_in_df, on=['bedrooms','room_type']).merge(types_bedrooms_location_df,on=['bedrooms','room_type'])
review_df = pd.merge(first_merge,second_merge, on=['bedrooms','room_type']).merge(types_bedrooms_ratings_df,on=['bedrooms','room_type'])
review_df
```

```python
# Narrow Dataframe into 4 Groups

review_filtered_df = review_df.iloc[0:8,:]
review_filtered_df
review_filtered_studio = review_df.iloc[14:16,:]
review_filtered_studio
df_list = [review_filtered_df, review_filtered_studio]
review_merge = pd.concat(df_list)
review_merged = review_merge.rename =['review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_value',
                       'review_scores_communication', 'review_scores_location', 'review_scores_checkin',
                       'review_scores_rating']
review_merge
```

```python
# Dropped N/A values for plotting

large_host_drop = large_host_ratings.dropna(subset=['review_scores_rating'])
small_host_drop = small_host_ratings.dropna(subset=['review_scores_rating'])
single_host_drop = small_host_ratings.dropna(subset=['review_scores_rating'])
large_host = large_host_drop["review_scores_rating"]
small_host = small_host_drop["review_scores_rating"]
single_host = small_host_drop["review_scores_rating"]
```

```python
# Plot Historgram to Check for Normality of Hosts with more than 5 Listings

plt.hist(large_host, edgecolor='black', bins=20)
```

```python
# Shapiro Test to check for Normality

shapiro(large_host)
```

```python
# Shapiro Test to check for Normality

shapiro(single_host)
```

```python
# Ttest of Significance Population

# Perform t-test between single_host and small_host
result1 = stats.ttest_ind(single_host, small_host, equal_var=False)

# Perform t-test between single_host and large_host
result2 = stats.ttest_ind(single_host, large_host, equal_var=False)

# Perform t-test between small_host and large_host
result3 = stats.ttest_ind(small_host, large_host, equal_var=False)

# Print the results
print("T-test result for single_host and small_host:", result1)
print("T-test result for single_host and large_host:", result2)
print("T-test result for small_host and large_host:", result3)
 
```

```python
# Collect 50 Sets of Random Samples from Population and Calculate Mean of Each set(Host with Single Listing)

sample_size = 50
num_sets = 50

# List to store the sampled DataFrames
sampled_sets = []
sampled_set_means_single = []

# Loop to generate 50 sets of random samples
for _ in range(num_sets):
    sampled_data = single_host.sample(n=sample_size, random_state=42)
    sampled_sets.append(sampled_data)

# Loop to calculate the mean of each sampled set
for i in range(num_sets):
    mean = sampled_sets[i].mean()
    sampled_set_means_single.append(mean)

# Display the first few rows of the first sampled set and its mean
sampled_set_means_single[0]
```

```python
# Collect 50 Sets of Random Samples from Population and Calculate Mean of Each set(Host with Few Listings)

sample_size = 50
num_sets = 50

# List to store the sampled DataFrames
sampled_sets = []
sampled_set_means_small = []

# Loop to generate 50 sets of random samples
for _ in range(num_sets):
    sampled_data = small_host.sample(n=sample_size, random_state=42)
    sampled_sets.append(sampled_data)

# Loop to calculate the mean of each sampled set
for i in range(num_sets):
    mean = sampled_sets[i].mean()
    sampled_set_means_small.append(mean)

# Display the first few rows of the first sampled set and its mean
sampled_set_means_small[0]
```

```python
# Collect 50 Sets of Random Samples from Population and Calculate Mean of Each set(Host with Many Listings)

sample_size = 50
num_sets = 50

# List to store the sampled DataFrames
sampled_sets = []
sampled_set_means_large = []

# Loop to generate 50 sets of random samples
for _ in range(num_sets):
    sampled_data = large_host.sample(n=sample_size, random_state=42)
    sampled_sets.append(sampled_data)

# Loop to calculate the mean of each sampled set
for i in range(num_sets):
    mean = sampled_sets[i].mean()
    sampled_set_means_large.append(mean)

# Display the first few rows of the first sampled set and its mean
sampled_set_means_large[0]
```

```python
# Ttest of Significance for Sampled Data of Population

# Perform t-tests
t_statistic_single_small, p_value_single_small = stats.ttest_ind(sampled_set_means_single, sampled_set_means_small)
t_statistic_single_large, p_value_single_large = stats.ttest_ind(sampled_set_means_single, sampled_set_means_large)
t_statistic_small_large, p_value_small_large = stats.ttest_ind(sampled_set_means_small, sampled_set_means_large)

# List the t-statistic and p-value for each test
print("t-statistic (Single vs Small):", t_statistic_single_small)
print("p-value (Single vs Small):", p_value_single_small)

print("t-statistic (Single vs Large):", t_statistic_single_large)
print("p-value (Single vs Large):", p_value_single_large)

print("t-statistic (Small vs Large):", t_statistic_small_large)
print("p-value (Small vs Large):", p_value_small_large)

# Check the p-values to determine significance
if p_value_single_small < 0.05:
    print("Single vs Small: There is a significant difference.")
else:
    print("Single vs Small: There is no significant difference.")

if p_value_single_large < 0.05:
    print("Single vs Large: There is a significant difference.")
else:
    print("Single vs Large: There is no significant difference.")

if p_value_small_large < 0.05:
    print("Small vs Large: There is a significant difference.")
else:
    print("Small vs Large: There is no significant difference.")
```
